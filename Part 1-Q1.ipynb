Q1. It is a general belief that studying at the main campus will give the students better work opportunities and leave them with less debt. Is this true?

```
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import r2_score,mean_squared_error
import matplotlib.pyplot as plt
```

My first model will have the main campus feature, median eranings and the national rank of the university as the inputs.

```
X=college_df[['MAIN','MD_EARN_WNE', 'national_rank']]
```

```
#Setting the target variable
Y=college_df['DEBTMEDIAN']
```

```
# Split the data into Trainging and Testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size = 0.30,random_state =42)
```

```
svr_model=SVR()
```

```
svr_model.fit(X_train,Y_train)
```

```
y_pred=svr_model.predict(X_test)
y_pred
```

I will be training my second model without using the main campus feature.

```
X_nomain=college_df[['MD_EARN_WNE', 'national_rank']]
```

```
Y_nomain=college_df['DEBTMEDIAN']
```

```
# Split the data into Trainging and Testing sets
X_train_nomain, X_test_nomain, Y_train_nomain, Y_test_nomain = train_test_split(X_nomain, Y_nomain,test_size = 0.30,random_state =42)
```

```
svr_model_nomain=SVR()
```

```
svr_model_nomain.fit(X_train_nomain,Y_train_nomain)
```

```
y_pred_nomain=svr_model_nomain.predict(X_test_nomain)
y_pred_nomain
```

Evaluating models using the mean squared error. The mean squared error tells you how close a regression line is to a set of points.

```
#Model evaluation: main campus feature included
mse=mean_squared_error(Y_test,y_pred)
rsme=np.sqrt(mse)
rsme
```

```
#Model evaluation: without main campus feature
mse_nomain=mean_squared_error(Y_test_nomain,y_pred_nomain)
rsme_nomain=np.sqrt(mse_nomain)
rsme_nomain
```

The larger the number the larger the error. Error in this case is the difference between the observed values and the predicted ones.

```
from sklearn.metrics import r2_score
```

```
r2_score(Y_test, y_pred)
```

```
r2_score(Y_test_nomain, y_pred_nomain)
```

R score is negative.. the model is arbitrarily bad
